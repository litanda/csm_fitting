# To fit mode to the data with the kernel, we will need a mean function. 

- The mean function represent the model systematic errors, which is identified as the frequency offsets between observation and the best-fitting model. However, the best-fitting model is unknow before doting the fits. For this reason, we use two priors to determine the mean funciton. The detials of the method can be seen in the paper.

# To consider the misreported mode, we need to apply the multivariate t-distribution instead of the multivariate normal distribution. 

- the t-distribution is basicly a normal distribution with long tails. The tails represent the chance to have an outlier in a population.    

```python
# Here we update our code to involve the mean function and a likeliood function based on the t-distribution

class smart_fit:
    '''
    
    '''
    
    def __init__(self):
        return None
    
    def mle(self, model, obs, obs_e):
        return -0.5*(model - obs)**2.0/obs_e**2.0
    
    def fit_modes(self, obs_nu, obs_nu_e, modeldf, nucolumns, lkcolumn):
        
        if lkcolumn in modeldf.columns:
            modeldf.loc[:,lkcolumn] = -1.0e99
        else:    
            modeldf.insert(loc = len(modeldf.columns), column = lkcolumn, value = -1.0e99,allow_duplicates=False) 
        
        for index, row in modeldf.iterrows():
            loglk = 0.0            
            model = row[nucolumns]
            loglk = self.mle(model, obs_nu, obs_nu_e)
            modeldf.loc[index, lkcolumn] = np.sum(loglk)
        return modeldf
    
#   fit with covirance function

#   fit with covirance function
    
    def fit_modes_cov(self, obs_nu = None, obs_nu_e = None, n_order = None,obs_dnu = None,
                       mean_sys = None,
                       sigma_e = None, length_e = None, 
                       sigma_u1 = None, sigma_u1_mlt = None, length_u1 = None,
                       sigma_u2 = None, length_u2 = None, 
                       modeldf = None, nucolumns = None, lkcolumn = None, dis ='normal'):
        
        if lkcolumn in modeldf.columns:
            modeldf.loc[:,lkcolumn] = -1.0e99
        else:    
            modeldf.insert(loc = len(modeldf.columns), column = lkcolumn, value = -1.0e99,allow_duplicates=False) 
        
        def SEKernel(x, y, sigma, length):
            return sigma**2 * np.exp(-0.5 * np.subtract.outer(x, y)**2 / length**2)

        def cov_loglk(model, obs, obs_e, n_order):
            
#           We have four kernels to present the noises in 
#            i) model errors(the surface term), 
#           ii) frequency changes between grid for mass, matellicity, and helium faction 
#          iii) frequency changes between grid for the mixing-length parameter 
#           iv) changes in the glitch signiture between grid.

            Sigma = SEKernel(obs, obs, sigma_e, length_e) \
                  + SEKernel(obs, obs, sigma_u1, length_u1) \
                  + SEKernel(obs, obs, sigma_u1_mlt, length_u1) \
                  + SEKernel(obs, obs, sigma_u2, length_u2)
            
            mean_function = mean_sys
            freq_dif = model + mean_function - obs
            
            if dis == 'normal': 
                loglk = -0.5 * freq_dif.dot( np.linalg.inv(Sigma + np.identity(len(n_order))*obs_e**2.0)).dot(freq_dif.T) 
            elif dis == 't':
                df = 2
                p = len(n_order)
                loglk = np.log((1.0 + 1/df*freq_dif.dot( np.linalg.inv(Sigma + np.identity(len(n_order))*obs_e**2.0)).dot(freq_dif.T))**(-(df+p)/2))
        
            else: 
                print("Distribution is either 'normal' or 't'!")
                stop
            return loglk

        '''
        '''
        for index, row in modeldf.iterrows():
                
            model_nu = row[nucolumns]
     
            modeldf.loc[index, lkcolumn] = cov_loglk(model_nu, obs_nu, obs_nu_e, n_order)
        return modeldf
    
    
    
    def flattop_gaussian(self, x, mu, sigma):
        return -( (x-mu)**2/2/sigma**2 )**10.0
    
    def gaussian(self, x, mu, sigma):
        return -( (x-mu)**2/2/sigma**2 )
    
#   Determine the mean function: an intitail estimate of the model errors
    def mean_function(self, obs_nu = None, obs_numax = None,
                      model_df = None, nucolumns = None, frac = 0.3,
                      mu_lownu = 0.0, sigma_lownu = 0.001,
                       mu_numax = None, sigma_numax = None,
                       plot = None, ax = None, cmap = None, norm = None):
        
        newcolumn = 'lk_mean'
        df = model_df.copy()
        
        if frac < 1.0:
            df = df.sample(frac = frac, random_state = 999)
            df.index = range(len(df))
        
        df.insert(0, newcolumn, value = -1.0e99)
        
        fitx = fity = fitlk = np.array([])
        
        for index, row in df.iterrows():
            model_nu = row[nucolumns].to_numpy(dtype=float)
            
            dnu_sc = obs_nu - model_nu
            dnu_lownu = dnu_sc[obs_nu<=0.7*obs_numax]
            
            lk_lownu = self.gaussian(np.median(dnu_lownu)/obs_numax, mu_lownu, sigma_lownu)
            lk_numax = self.flattop_gaussian(np.median(dnu_sc)/obs_numax, mu_numax, sigma_numax)
            
            if np.median(dnu_sc) >= 0.0:
                lk_numax = -1.0e99
            
            df.loc[index,newcolumn] = lk_lownu + lk_numax
            
            lk = np.full(len(obs_nu), np.exp(lk_lownu + lk_numax))
            
        
            fitx = np.concatenate((fitx, obs_nu))
            fity = np.concatenate((fity, dnu_sc))
            fitlk = np.concatenate((fitlk, lk))
                        
            if ((plot == True) & (np.exp(lk_lownu+lk_numax)>= 0.1)): 
                ax.scatter(obs_nu, dnu_sc,s = 5, alpha = 0.3, color = cmap(norm(np.exp(lk_lownu+lk_numax))))
            
        fitp,cov = np.polyfit(fitx, fity, 3, w = fitlk,cov = True) 
        
        mean_sys = fitp[0]*obs_nu**3.0 + fitp[1]*obs_nu**2.0 +fitp[2]*obs_nu**1.0 +fitp[3]
                
        variance = np.average( (obs_nu - df[nucolumns] - mean_sys)**2, weights = np.exp(df[newcolumn]),  axis = 0)
        
        return df, mean_sys, np.sqrt(variance)
    
sfit = smart_fit()
```
